```python
import pandas as pd

df = pd.read_csv('test-sales.csv')
months = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

df = df.assign(month=months)
df.set_index('month', inplace=True)
dados = pd.DataFrame({ 'months': df.index.tolist(), 'revenue': df['revenue'].tolist(), 'quantity': df['units'].tolist() })
dados.set_index('months', inplace=True)
dados

import matplotlib.pyplot as plt

plt.figure(figsize=(24, 6), facecolor='#f5f5f4')
plt.plot(dados.index, dados['revenue'], label='Revenue', color="#2EC4B6")
plt.xticks(rotation=45)
```
